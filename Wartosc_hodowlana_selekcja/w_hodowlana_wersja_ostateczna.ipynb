# $$ \stackrel\frown{G} = b_{yp} (P_i - \bar{P}) + \bar{P} $$

```python
import matplotlib.pyplot as plt

kolumny = dane.columns
srednie = dane.groupby(['ferma']).mean()

for i in range(7,len(kolumny)):
    wlk = dane[['rok','ferma','nr os','nr oj','nr ma',kolumny[i]]]
    for number in range(0,7):
        wlk['byp'] = wsp_reg['h^2_dobre'][number]
    wlk['mean'] = 0
    wlk['mean'] = wlk.apply(lambda row: srednie[kolumny[i]].iloc[0] 
                        if row['ferma'] == 1 else (srednie[kolumny[i]].iloc[1] 
                        if row['ferma'] == 2 else (srednie[kolumny[i]].iloc[2] 
                        if row['ferma'] == 3 else '0')), axis = 1)
    #print(wlk.head())
    wlk = wlk.assign(G=lambda x: (x['byp'] *(x[kolumny[i]] - x['mean']) + x['mean']))
    
    diff = wlk[['rok','ferma','G']]
    
    diff = diff.groupby(['ferma','rok']).mean()
    
    wlk['G'].round(2).value_counts().plot(kind='barh')
    plt.title("Rozklad wartosci dla wartosci hodowlanej dla " + kolumny[i])
    plt.grid()
    plt.show()
    
    fig, ax = plt.subplots(figsize=(15,7))
    wlk.groupby(['rok','ferma']).mean()['G'].unstack().plot(ax=ax)
    plt.title("Wartosc hodowlana na przestrzeni lat dla " + kolumny[i])
    plt.grid()
    plt.show()
```

# Selekcja

$$ ∆A = (A_1 - A_0)$$
$$ A_1 = {A_o + A_m \over 2}$$

```python
kolumny = dane.columns
srednie = dane.groupby(['ferma']).mean()

for i in range(7,len(kolumny)):
    wlk = dane[['rok','ferma','nr os','nr oj','nr ma',kolumny[i]]]
    for number in range(0,7):
        wlk['byp'] = wsp_reg['h^2_dobre'][number]
    wlk['mean'] = 0
    wlk['mean'] = wlk.apply(lambda row: srednie[kolumny[i]].iloc[0] 
                        if row['ferma'] == 1 else (srednie[kolumny[i]].iloc[1] 
                        if row['ferma'] == 2 else (srednie[kolumny[i]].iloc[2] 
                        if row['ferma'] == 3 else '0')), axis = 1)
    #print(wlk.head())
    wlk = wlk.assign(G=lambda x: (x['byp'] *(x[kolumny[i]] - x['mean']) + x['mean']))
    wlk['G_ma' + kolumny[i]] = 0
    wlk['G_oj' + kolumny[i]] = 0
    
    for index, row in wlk.iterrows():
        nr_ojciec = wlk['nr oj'][index]
        g_oj = wlk['G'][index]
        for index2, row2 in wlk.iterrows():
            if nr_ojciec == wlk['nr os'][index2]:
                wlk['G_oj' + kolumny[i]][index2] = g_oj
                break
            else:
                continue
            
    for index, row in wlk.iterrows():
        nr_matka = wlk['nr ma'][index]
        g_ma = wlk['G'][index]
        for index2, row2 in wlk.iterrows():
            if nr_matka == wlk['nr os'][index2]:
                wlk['G_ma' + kolumny[i]][index2] = g_ma
                break
            else:
                continue
    break
```
